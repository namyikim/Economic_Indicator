```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader as pdr

plt.rcParams["figure.figsize"] = (20,6)
plt.rcParams["font.family"] = 'nanummyeongjo'
plt.rcParams["axes.grid"] = True
plt.rcParams["figure.figsize"] = (12,6)
plt.rcParams["axes.formatter.useoffset"] = False
plt.rcParams["axes.unicode_minus"] = False
```

```python
#원달라 환율
USD_KRX = fdr.DataReader('USD/KRW',start='2020-01-01')
graph = USD_KRX.loc[:,'Close'].plot()
graph.axhline(1050,ls='--',color='r')
graph.axhline(1150,ls='--',color='r')
```

```python
#KOSPI와 역순 원달라의 상관관계
USD_KRX = fdr.DataReader('USD/KRW',start='2020-01-01')
KOSPI = fdr.DataReader('KS11',start='2020-01-01')
concat_data = pd.concat([-USD_KRX['Close'], KOSPI['Close']], axis=1)
concat_data.columns = ['USD_KRX','KOSPI']
print(concat_data.corr()) #상관계수
concat_data.plot(secondary_y='KOSPI')

```

```python
# 삼성전자(005930) 와 KOSPI 상관관계
KOSPI = fdr.DataReader('KS11',start='2020-01-01')
Samsung = fdr.DataReader('005930','2020-01-01')
concat_data = pd.concat([Samsung['Close'], KOSPI['Close']], axis=1)
concat_data.columns = ['Samsung','KOSPI']
print(concat_data.corr()) #상관계수
concat_data.plot(secondary_y='KOSPI')
```

```python
#Samsung 주가와와 역순 원달라의 상관관계
USD_KRX = fdr.DataReader('USD/KRW',start='2020-01-01')
Samsung = fdr.DataReader('005930','2020-01-01')
concat_data = pd.concat([-USD_KRX['Close'], Samsung['Close']], axis=1)
concat_data.columns = ['USD_KRX','Samsung']
print(concat_data.corr()) #상관계수
concat_data.plot(secondary_y='Samsung')
```
